Проект: Оптимизация Доставки Интернет-Магазина

 1. Введение и Обзор Проблемы

О чём бизнес?
Интернет-магазин сталкивается с рядом операционных проблем, которые негативно влияют на качество обслуживания клиентов:

Задержки в доставке заказов.

Нестабильное качество работы курьеров.

Недовольство клиентов (низкий уровень Customer Satisfaction).

Цель бизнеса — улучшить процесс доставки и сервис для повышения лояльности и операционной эффективности.


    Цель проекта
Основная цель данного проекта — создать структурированную базу данных (БД).

Эта БД должна служить инструментом для глубокого анализа причин:

Возникновения задержек доставки.

Разработки мер по повышению удовлетворённости клиентов.


    Откуда данные?
Источником данных является история заказов из интернет-магазина. Эти данные включают важную информацию по следующим аспектам:

Состав покупки (что покупали).

Логистика доставки (как доставляли, из какого склада).

Обратная связь от клиентов (что думают клиенты).
